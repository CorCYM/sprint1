```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# 발전설비추이 -> generate
generate = pd.read_csv('data/발전설비추이.csv', encoding='cp949')
```

```python
# 발전설비추이 컬럼 확인
generate.columns
```

```python
# 발전설비추이 인덱스 중 (증감률) 행 전부 삭제
gradient = generate[generate['년  월'].str.contains('(증감률)')].index
generate.drop(gradient, inplace=True)
generate.head(3)
```

```python
# rows, columns 전환
generate = generate.T
```

```python
# 첫번째 행을 columns으로 지정
generate = generate.rename(columns=generate.iloc[0])
generate.head(2)
```

```python
# 첫번째 행 삭제
generate = generate.drop(generate.index[0])
```

```python
# index 리셋
generate.reset_index(inplace=True)
```

```python
# 컬럼명 '날짜'로 변경
generate.rename(columns={'index':'날짜'}, inplace=True)
```

```python
# 컬럼 중 '총발전설비' 삭제
generate.drop(labels=['총발전설비'], axis=1, inplace=True)
generate.head(2)
```

```python
# 2021.01 ~ 2021.12 해당 행 추출해서 새로운 데이터프레임 만들기
targets = ['2021.01', '2021.02', '2021.03', '2021.04', '2021.05', '2021.06',
                   '2021.07', '2021.08', '2021.09', '2021.1', '2021.11', '2021.12']
generate21 = generate.loc[generate['날짜'].isin(targets)]
generate21
```

```python
# 인덱스 21행 -> 날짜 2021.1을 2021.10으로 수정
generate21.loc[20] = ('2021.10','23,250','37,838','41,170','23,824','2,159','4,700','515')
```

```python
# 날짜 컬럼 . -> -로 수정
generate21['날짜'] = generate21['날짜'].str.replace('.','-')
```

```python
generate21.reset_index(drop=True, inplace=True)
```

```python
generate21
```

```python
# 2021발전설비추이 data폴더에 파일 저장
generate21.to_csv('data/2021발전설비추이.csv')
```
