```python
import pandas as pd
import numpy as np
```

```python
# 발전전력량추이 -> electric
electric = pd.read_csv('data/발전전력량추이.csv', encoding='cp949')
```

```python
# 발전설비추이 인덱스 중 (증감률) 행 전부 삭제
gradient_e = electric[electric['년  월'].str.contains('(증감률)')].index
electric.drop(gradient_e, inplace=True)
electric.head(3)
```

```python
# rows, columns 전환
electric = electric.T
```

```python
# 첫번째 행을 columns으로 지정
electric = electric.rename(columns=electric.iloc[0])
electric.head(2)
```

```python
# 첫번째 행 삭제
electric = electric.drop(electric.index[0])
```

```python
# index 리셋
electric.reset_index(inplace=True)
```

```python
# 컬럼명 '날짜'로 변경
electric.rename(columns={'index':'날짜'}, inplace=True)
```

```python
# 컬럼 중 '총발전량' 삭제
electric.drop(labels=['총발전량'], axis=1, inplace=True)
electric.head(2)
```

```python
# 2021.01 ~ 2021.12 해당 행 추출해서 새로운 데이터프레임 만들기
targets = ['2021.01', '2021.02', '2021.03', '2021.04', '2021.05', '2021.06',
                   '2021.07', '2021.08', '2021.09', '2021.1', '2021.11', '2021.12']
electric21 = electric.loc[electric['날짜'].isin(targets)]
electric21
```

```python
# 인덱스 21행 -> 날짜 2021.1을 2021.10으로 수정
electric21.loc[20] = ('2021.10','13,462','15,398','13,005','3,509','248','292','233')
```

```python
# 날짜 컬럼 . -> -로 수정
electric21['날짜'] = electric21['날짜'].str.replace('.','-')
```

```python
electric21.reset_index(drop=True, inplace=True)
```

```python
electric21
```

```python
# 2021발전전력량추이 data폴더에 파일 저장
electric21.to_csv('data/2021발전전력량추이.csv')
```
