```python
# 인덱스 변경
data = data.set_index('시점')
```

```python
# 최심신적설 빈 데이터 0으로 변경
data['최심신적설 (㎝)'] = data['최심신적설 (㎝)'].str.replace('-','0')
```

```python
# 데이터 전체 실수값으로 변경
data = data.astype(float)
```

```python
# 날씨 요소들 간 상간계수 테이블
mask = np.triu(np.ones_like(data.corr()))
plt.figure(figsize=(20,10))
sns.heatmap(data.corr(),fmt='.2f', cmap='Reds',mask=mask, annot=True, )
```

```python
# 상관계수가 높은 컬럼만 뽑아서 결과 확인
f = ['평균 기온 (℃)','강수량 (mm)','평균 상대습도 (%)','이슬점온도 (℃)', '평균운량 (10%)','최대풍속 (m/s)']
fig , axs = plt.subplots(figsize=(16,8),ncols=3,nrows=2)
for i, f in enumerate(f):
    row = int(i/3)ㄱ
    col = i%3
    sns.regplot(x=f, y='평균전력수요량 (kWh)', data=data, ax = axs[row][col])
```

```python
# 각 그래프 상관계수 출력
f = ['평균 기온 (℃)','강수량 (mm)','평균 상대습도 (%)','이슬점온도 (℃)', '평균운량 (10%)','최대풍속 (m/s)']
for i in f:
    print(np.round(data[i].corr(data['평균전력수요량 (kWh)']),3))
```
